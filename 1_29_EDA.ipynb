## 3.Univariate Analysis (Numerical Features)(Histogram, Box Plot)

```python
df['salary'].describe()
```

```python
plt.hist(df['salary'], bins=6)
plt.title('Histogram')
plt.show()
```

```python
plt.boxplot(df['salary'])
plt.title('Boxplot')
plt.show()
```

## 4.Univariate Analysis (Categorical Features)

```python
df.columns
```

```python
df['job_title'].value_counts().plot(kind='barh')
plt.show()
```

## 5. Bivariate Analysis (Numerical vs Numerical)

```python
plt.scatter(df['experience_level'], df['salary'])
plt.show()
```

```python
plt.bar(df['experience_level'], df['salary'])
plt.show()
```

## 6. Bivariate Analysis (Categorical vs Numerical)

```python
plt.scatter( df['work_year'], df['salary'],color='orange')
plt.show()
```

## 7. Multivariate Analysis (Correlation Heatmap, Pair Plot)

```python
sns.pairplot(df)
plt.title('Pair Plot')
plt.show()
```

```python
d_h = np.random.randint(1,200, (10, 10)) 
sns.heatmap(d_h)
plt.title('Heatmap')
plt.show()
```
